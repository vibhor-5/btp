```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, SimpleRNN, Dense
from tensorflow.keras.optimizers import Adam

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python
if 'combined_df' not in locals():
    combined_df = pd.read_parquet("combined_dataset.parquet")

combined_df = combined_df.copy()

combined_df["dt"] = pd.to_datetime(combined_df["dt"])
```

```python
combined_df.head()
```

```python
combined_df["daily_stockout"] = combined_df["hours_stock_status"].apply(
    lambda x: int(np.any(np.array(x) == 1))
)
combined_df[["hours_stock_status","daily_stockout"]].head()
```

```python
combined_df["avg_hourly_sales"] = combined_df["hours_sale"].apply(np.mean)

combined_df["max_hourly_sales"] = combined_df["hours_sale"].apply(np.max)

combined_df["std_hourly_sales"] = combined_df["hours_sale"].apply(np.std)
```

```python
feature_cols = [

    # Product hierarchy
    "city_id",
    "management_group_id",
    "first_category_id",
    "second_category_id",
    "third_category_id",

    # Sales
    "sale_amount",
    "avg_hourly_sales",
    "max_hourly_sales",
    "std_hourly_sales",

    # Inventory
    "stock_hour6_22_cnt",

    # Pricing
    "discount",

    # Calendar
    "holiday_flag",
    "activity_flag",

    # Weather
    "precpt",
    "avg_temperature",
    "avg_humidity",
    "avg_wind_level"
]


target = "daily_stockout"
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

combined_df[feature_cols] = scaler.fit_transform(
    combined_df[feature_cols]
)
```

```python
LOOKBACK = 7
TRAIN_DAYS = 75
```

```python
X_train = []
y_train = []

X_test = []
y_test = []

for (_, _), group in combined_df.groupby(["store_id", "product_id"]):

    group = group.sort_values("dt").reset_index(drop=True)

    if len(group) < 90:
        continue

    features = group[feature_cols].values
    labels = group[target].values

    # ---------- Training ----------
    for i in range(LOOKBACK, TRAIN_DAYS):

        X_train.append(features[i-LOOKBACK:i])
        y_train.append(labels[i])

    # ---------- Testing ----------
    for i in range(TRAIN_DAYS, len(group)):

        X_test.append(features[i-LOOKBACK:i])
        y_test.append(labels[i])
```

```python
X_train = np.array(X_train, dtype=np.float32)
y_train = np.array(y_train, dtype=np.float32)

X_test = np.array(X_test, dtype=np.float32)
y_test = np.array(y_test, dtype=np.float32)
```

```python
print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

```python
model = Sequential([
    Input(shape=(LOOKBACK, X_train.shape[2])),
    SimpleRNN(32, activation="tanh"),
    Dense(1, activation="sigmoid")
])
```

```python
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=20,
    batch_size=64,
    verbose=1
)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)

print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```
